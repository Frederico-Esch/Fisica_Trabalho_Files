```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("data.csv")
data
```

```python
t_mean = []
t_std  = []

y_mean = []
y_std  = []
for i in data.iterrows():
    i = i[1].values
    
    t_mean.append(np.mean([i[0], i[2], i[4]]))
    t_std.append(np.std([i[0], i[2], i[4]]))
    
    y_mean.append(np.mean([i[1], i[3], i[5]]))
    y_std.append(np.std([i[1], i[3], i[5]]))

data['t_mean'] = t_mean
data['y_mean'] = y_mean

data['t_std'] = t_std
data['y_std'] = y_std
```

```python
data
```

```python
data.to_csv("dataset.csv", index=0)
```
